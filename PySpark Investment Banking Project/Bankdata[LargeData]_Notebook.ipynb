# to get all the columns missing values at single instance

```python
#to get all the columns missing values at single instance
def check_miss_value_pctg(data,lst_cl):
    missing_values = {}
    for i in lst_cl:
        a = data.filter(col(i).isNull()).count()
        missing_values[i] = a
    return (missing_values)
```

```python
check_miss_value_pctg(df,df.columns)
```

# to get all the columns missing values percentage at single instance

```python
#to get all the columns missing values percentage at single instance
def check_missing_value_pcntg(data,lst_cl):
    missing_values_less_than_75 = {}
    missing_values_greater_than_75 = {}
    b = data.count()
    for i in lst_cl:
        a = df.filter(col(i).isNull()).count()
        c = (a/b) * 100
        if c >= 75:
            missing_values_greater_than_75[i] = c
        else:
            missing_values_less_than_75[i] = c
    return ({"Missing_Value_above75:":missing_values_greater_than_75,"Missing_Value_below75:":missing_values_less_than_75})
```

```python
check_missing_value_pcntg(df,df.columns) #checking the missing values percentage of each columns
```

```python
miss = check_missing_value_pcntg(df,df.columns) #assigning to df
```

```python
column_names = list(miss['Missing_Value_above75:'].keys())
print("columnNames:",column_names)                             #extratcing columns that nulls more than 75%  
```

# To chec duplicates 

```python
def check_dups(data,cl):
    a = data.select(cl).distinct().count()
    b = data.count()
    if a == b:
        print("No Duplicates")
    else:
        c = b - a
        print("There are", c, "Duplicates")
```

```python
check_dups(df,"id")
```

```python
check_dups(df,"num_bc_tl")
```

# To check all column Duplicates at single instance

```python
def duplicate_check_allclms(data):
    for cl in df.columns:                            #writing fro loop
        a = data.select(cl).distinct().count()     #define variable
        b = data.count()                             #define variable
        if a == b:
            print(f"No Duplicates in column: {cl}")
        else:
            c = b-a
            print(f"There are {c} duplicates in column: {cl}")
```

```python
duplicate_check_allclms(df)
```
