# Now I can calculate all the columns missing values at single instance with Functions

```python
def check_miss_values_count (data, lst_cl):
    missing_values = {}
    for i in lst_cl:
        a = data.filter(col(i).isNull()).count()
        missing_values[i] = a
    return (missing_values) 
```

```python
check_miss_values_count(df,df.columns)
```

# To get the missing value percentage with Functions

```python
def check_miss_values_pct(data,lst_cl):
    miss_value_pct = {}
    for i in lst_cl:
        a = data.filter(col(i).isNull()).count()
        b = data.count()
        c = (a/b) * 100
        miss_value_pct[i] = c
    return (miss_value_pct)
```

```python
check_miss_values_pct(df,df.columns)
```

# To check distinct values of all columns by using function

```python
def check_dist_values(data,lst_cl):
    dist_values_count = {}
    for i in lst_cl:
        a = data.select(i).distinct().count()
        dist_values_count[i] = a
    return (dist_values_count)
```

# calculating a function for loya customer category

```python
def loyalcustcate(cust_firm_sales_percen):
    if cust_firm_sales_percen == 100:
        return "loyal customer"
    elif cust_firm_sales_percen > 75 and cust_firm_sales_percen < 100:
        return "top-tier loyal customer"
    elif cust_firm_sales_percen > 50 and cust_firm_sales_percen <=75:
        return "mid-tier loyal customer"
    elif cust_firm_sales_percen > 25 and cust_firm_sales_percen <=50:
        return "low-tier loyal customer"
    elif cust_firm_sales_percen  > 0 and cust_firm_sales_percen <=25:
        return "basic loyal customer"
    else:
        return "stangant loyal customer"
```

```python
loyalcustcate(0)
```

# Now Applying this function to cust_firm_sales_percen and creating new column called custoemr_status

```python
cust_loyal_udf = udf(loyalcustcate,StringType())
df = df.withColumn("customer_category",cust_loyal_udf(df['cust_firm_sales_percen']))
```
